```python
import pandas as pd
```

```python
#Read data from Seto Lab 
data_0 = pd.read_csv("AllPop.csv")
data_0.head()
```

```python
data_0 = data_0.iloc[:,[0,2,3,6,4]]
data_0.head()
```

```python
#Read data from the World Bank
data_1 = pd.read_csv("pop1980&2015_update.csv")
data_1.head()
```

```python
data_1980 = data_1.iloc[:,[1,2,3,4]]
data_1980['Year'] = 1980
data_1980.head(3)
```

```python
data_1980 = data_1980.rename(columns={'Urban':'City','1980': 'Value'})
data_1980.head(3)
```

```python
data_2015 = data_1.iloc[:,[1,2,3,5]]
data_2015['Year'] = 2015
data_2015.head(3)
```

```python
data_2015 = data_2015.rename(columns={'Urban':'City','2015': 'Value'})
data_2015.head(3)
```

```python
#Combine data altogether
df = data_0
df = data_0.append(data_1980)
df = df.append(data_2015)
df.head(3)
```

```python
len(df)
```

```python
#Normalise data for visualisation 
df["VAL"] = (df["Value"] - df["Value"].min()) / (df["Value"].max() - df["Value"].min())
df.head()
```

```python
df.describe()
```

```python
#Export data to csv
df.to_csv("GlobalPopHistory.csv")
```
